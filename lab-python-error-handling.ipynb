def initialize_inventory(products):
    inventory = {}
    for product in products:
        valid_quantity = False
        while not valid_quantity:
            try:
                quantity = int(input(f"Enter the quantity of {product}s available: "))
                if quantity < 0:
                    raise ValueError("Invalid quantity! Please enter a non-negative value.")
                valid_quantity = True
            except ValueError as error:
                print(f"Error: {error}")
        inventory[product] = quantity
    return inventory

def initialize_inventory(products):

                    inventory[product] = quantity
                    valid_input = True
                else:
                    print("Quantity cannot be negative. Please enter a valid quantity.")

````python
#3
def get_customer_orders(n_of_orders):
    customer_orders = set()  # Initialize a set to collect unique product names
    for _ in range(n_of_orders):
        valid_input = False
        while not valid_input:
            try:
                product_name = input("Enter product name: ")
                if not product_name.strip():
                    raise ValueError("Product name cannot be empty.")
                customer_orders.add(product_name)  # Add to the set
                valid_input = True
            except ValueError as error:
                print(f"Error: {error}")
    return customer_orders

try:
    n_of_orders = int(input("Enter the number of customer orders: "))
    if n_of_orders <= 0:
        raise ValueError("The number of customer orders should be a positive integer.")
    
    customer_orders = get_customer_orders(n_of_orders)
    print("customer_orders are:", customer_orders)
except ValueError as error:
    print(f"Error: {error}. Please restart the program and enter a valid number.")
````

````python
#4

inventory  = ["t-shirt", "mug", "hat", "book", "keychain"]

def get_customer_orders(n_of_orders, inventory):
    customer_orders = set()  # Initialize a set to collect unique product names
    for _ in range(n_of_orders):
        valid_input = False
        while not valid_input:
            try:
                product_name = input("Enter product name that is available")
                if product_name not in inventory:
                    raise ValueError("Product name is not available .")
                customer_orders.add(product_name)  # Add to the set
                valid_input = True
            except ValueError as error:
                print(f"Error: {error}")
    return customer_orders

try:
    n_of_orders = int(input("Enter the number of customer orders: "))
    if n_of_orders <= 0:
        raise ValueError("The number of customer orders should be a positive integer.")
    
    customer_orders = get_customer_orders(n_of_orders, inventory)
    print("customer_orders are:", customer_orders)
except ValueError as error:
    print(f"Error: {error}. Please restart the program and enter a valid number.")
````
